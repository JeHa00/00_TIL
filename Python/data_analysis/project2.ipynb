```python
import matplotlib.pyplot as plt 
import pandas as pd 

cat_meal_data = {
    "브랜드": ["G사", "R사", "R사", "N사", "N사"], 
    "사료 이름": ["스킨앤코트", "어덜트 컴플리트", "어덜트 라이트", "어덜트 베이직", "어덜트 피쉬"], 
    "열량": [4122, 3967, 3752, 3872, 3988],
    "사료 크기":[8, 11, 10, 5, 5],
    "단백질 비율": [29, 32, 35, 31, 30], 
}

df = pd.DataFrame(cat_meal_data)

color_map = {"G사": "orange", "R사": "black", "N사": "brown"}


df["단백질 / 열량"] = df["단백질 비율"] / df["열량"]
df["열량 / 사료크기"] = (df["열량"] / df["사료 크기"]).round()

plt.figure(figsize=(4, 5))
plt.rc("font", family="AppleGothic")
```

```python
for index, row in df.iterrows():
    color = color_map[row["브랜드"]]

    plt.scatter(row["단백질 / 열량"], row["열량 / 사료크기"], s=row["사료 크기"] * 100, alpha=0.6, color=color)

    plt.annotate(row["브랜드"], 
                 (row["단백질 / 열량"], row["열량 / 사료크기"]), 
                 textcoords="offset points", 
                 xytext=(10, 0), 
                 ha="left", 
                 fontsize=16
                )

plt.title("고양이 사료 정보", fontsize=20)
plt.xlabel("열량 대비 단백질 비율", fontsize=16)
plt.ylabel("사료 크기 대비 열량", fontsize=16)

plt.xlim(0.0065, 0.01)
plt.ylim(200, 900)

plt.grid(True)
plt.show()
```
