```python
import pandas as pd 

fight_score_df = pd.DataFrame({
    "경기시간": [
        "2021-04-20 09:15:00",
        "2022-02-15 15:30:00",
        "2023-04-17 20:45:00",
        "2024-06-25 10:31:00",
        "2023-06-25 10:32:00",
        "2022-05-20 22:33:00",
        "2024-06-25 10:31:00",
    ],
    "이름": ["오공","오반", "베지터", "오천", "브로리", "크리링", "트랭크스"],
    "전투력": [100000, 50000, 90000, 20000, 120000, 23000, 20000],
})

fight_score_df["경기시간"] = pd.to_datetime(fight_score_df["경기시간"])

fight_score_df
```

```python
fight_score_df["일"] = fight_score_df["경기시간"].dt.day
fight_score_df["시"] = fight_score_df["경기시간"].dt.hour
fight_score_df["분"] = fight_score_df["경기시간"].dt.minute

fight_score_df
```

```python
fight_score_df["이달의 날짜 수"] = fight_score_df["경기시간"].dt.days_in_month

fight_score_df
```

```python
fight_score_df["날짜"] = fight_score_df["경기시간"].dt.normalize()

fight_score_df
```

```python
fight_score_df["새 날짜 형식"] = fight_score_df["경기시간"].dt.strftime("%Y.%m.%d")

fight_score_df
```

```python
def assign_grade(fight_score):
    if fight_score > 90000:
        return "S"
    elif fight_score > 70000:
        return "A"
    elif fight_score > 50000:
        return "B"
    else:
        return 'C'

fight_score_df["전투력 등급"] = fight_score_df["전투력"].apply(assign_grade)

fight_score_df
```

```python
group_scores = fight_score_df.groupby("전투력 등급")

average_score = group_scores["전투력"].mean().rename("등글별 평균 전투력")

average_score
```

```python
amount = group_scores["전투력"].count().rename("등급별 수")

amount
```

```python
total_score = group_scores["전투력"].sum().rename("등급별 전투력 총합")

total_score
```
